 **Why is it popular?**

 *   **Easy to learn:**  Simple syntax makes it beginner-friendly.
 *   **Large community:** Extensive support and resources available.
 *   **Versatile:**  Can be used for a wide range of applications.

## Basic Syntax
### Print()
The **print()** function displays output to the console

```python
print("Hello, world!")
```

### Comments
Comments are used to explain code. Python use **#** for single-line comments.

```python
#This is a comment. It is ignored by the Python interpreter.
print("This line will be executed. ")
```

###Variables

Variables are used to **store data**. You assign a value to a variable using the **=** operator.

```python
message = "Welcome to Python!"  # String variable
number = 10                      # Integer variable
pi = 3.14159                    # Float variable
is_python_fun = True           # Boolean variable

print(message)
print(number)
print(pi)
print(is_python_fun)
```
