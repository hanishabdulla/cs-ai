 ### Understanding Data Types

 Python has several built-in data types, including:

 *   **Integers (int):** Whole numbers (e.g., 1, -5, 100).
 *   **Floats (float):** Decimal numbers (e.g., 3.14, -2.5, 0.0).
 *   **Strings (str):** Sequences of characters (e.g., "Hello", "Python").
 *   **Booleans (bool):** Represent truth values (True or False).

 Let's see some examples:
 ```python
 integer_number = 10
 float_number = 3.14
 string_text = "Python is fun!"
 boolean_value = True

 print(type(integer_number))
 print(type(float_number))
 print(type(string_text))
 print(type(boolean_value))

###Input and Output: Using input()

The input() function allows you to get input from the user. It always returns a string. You might need to convert the input to the appropriate data type (e.g., int() or float()).

````python
name = input("Enter your name: ")
print("Hello, " + name + "!")

age = input("Enter your age: ")
age = int(age)  # Convert the input to an integer
print("You are " + str(age) + " years old.") # Convert age to string for concatenation
````

###String Manipulation
####Concatenation

You can combine strings using the + operator.

````python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)
````

### String Methods
Python has many String methods, here are a few examples:

````python
text = "  Python is Awesome!  "

print(text.lower())   # Convert to lowercase
print(text.upper())   # Convert to uppercase
print(text.strip())   # Remove leading/trailing whitespace
print(text.replace("Awesome", "Great")) # Replace a substring
print(text.split(" "))  # Split the string into a list of words
````

### Example Code

````python
num1_str = input("Enter the first number: ")
num2_str = input("Enter the second number: ")

num1 = float(num1_str)
num2 = float(num2_str)

sum_result = num1 + num2
difference_result = num1 - num2
product_result = num1 * num2

# Avoid division by zero
if num2 != 0:
 quotient_result = num1 / num2
else:
 quotient_result = "Cannot divide by zero"

print(f"{num1} + {num2} = {sum_result}")
print(f"{num1} - {num2} = {difference_result}")
print(f"{num1} * {num2} = {product_result}")
print(f"{num1} / {num2} = {quotient_result}")

````
