 ### Arithmetic Operators

 Arithmetic operators perform mathematical calculations.

 | Operator | Description          | Example      |
 | :------- | :------------------- | :----------- |
 | `+`      | Addition             | `x + y`      |
 | `-`      | Subtraction          | `x - y`      |
 | `*`      | Multiplication       | `x * y`      |
 | `/`      | Division             | `x / y`      |
 | `//`     | Floor Division       | `x // y`     |
 | `%`      | Modulus (remainder)  | `x % y`      |
 | `**`     | Exponentiation       | `x ** y`     |

 Example:

```python
x = 10
 y = 3

 print(f"x + y = {x + y}")       # Addition
 print(f"x - y = {x - y}")       # Subtraction
 print(f"x * y = {x * y}")       # Multiplication
 print(f"x / y = {x / y}")       # Division
 print(f"x // y = {x // y}")     # Floor Division
 print(f"x % y = {x % y}")       # Modulus
 print(f"x ** y = {x ** y}")     # Exponentiation
```

### Comparison Operators
Comparison operators compare two values. They return a boolean value (**True** or **False**).

---

Screenshot 2025-04-07 at 13.42.03.jpg

Example:

```python
x = 5
y = 10

print(f"x == y: {x == y}")  # Equal to
print(f"x != y: {x != y}")  # Not equal to
print(f"x > y: {x > y}")   # Greater than
print(f"x < y: {x < y}")   # Less than
```

### Logical Operators
Logical operators combine boolean expressions

Operator | Description                                | Example
---------|--------------------------------------------|---------
`and`    | Returns True if both operands are True     | `x and y`
`or`     | Returns True if at least one operand is True | `x or y`
`not`    | Returns the opposite of the operand        | `not x`


### Operator Precedence
Operator precedence determines the order in which operations are performed. Use parentheses to override the default precedence. A common mneomnic is PEDMAS/BODMAS:



1.   Parentheses
2.   Exponents
3.   Multiplication and Division
4.   Addition and Subtraction

Example:



```python
result = 10 + 2 * 5  # Multiplication before addition
print(result)

result = (10 + 2) * 5  # Parentheses change the order
print(result)
```
